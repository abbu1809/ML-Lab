'r' → Red color

'^' → Triangle-up markers for each point

'-' → Solid line connecting the points

### Additional Visualization Examples

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Multi-line plot example
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--')
plt.plot(x, y3, label='sin(x)*cos(x)', color='green', linestyle=':')
plt.title('Multi-line Plot of Trigonometric Functions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Heatmap example
# Generate some random data for the heatmap
data = np.random.rand(10, 12)

plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Heatmap of Random Data')
plt.xlabel('Column Index')
plt.ylabel('Row Index')
plt.show()
```

```python
# Bar plot example
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 45, 56, 12]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Bar Plot of Categorical Data')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Histogram example
# Generate random data for the histogram
hist_data = np.random.normal(loc=0, scale=1, size=1000) # Mean 0, Std Dev 1

plt.figure(figsize=(8, 6))
plt.hist(hist_data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### More Advanced Visualization Examples

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Box Plot Example

# Generate some sample data for box plots
data_a = np.random.normal(0, 1, 100)
data_b = np.random.normal(1, 1.5, 100)
data_c = np.random.normal(0.5, 0.7, 100)

plt.figure(figsize=(10, 6))
sns.boxplot(data=[data_a, data_b, data_c], palette='pastel')
plt.xticks([0, 1, 2], ['Group A', 'Group B', 'Group C'])
plt.title('Box Plot of Sample Data Groups')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# 3D Scatter Plot Example

# Generate random data for 3D scatter plot
x_3d = np.random.rand(100)
y_3d = np.random.rand(100)
z_3d = np.random.rand(100)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_3d, y_3d, z_3d, c=z_3d, cmap='viridis', s=50)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Simple 3D Scatter Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parametric Plot Example (2D Spiral)

# Define the parameter t
t = np.linspace(0, 10 * np.pi, 500) # From 0 to 10*pi with 500 points

# Define parametric equations for a spiral
x_param = t * np.cos(t)
y_param = t * np.sin(t)

plt.figure(figsize=(8, 8))
plt.plot(x_param, y_param, color='purple', linestyle='-', linewidth=2)
plt.title('Parametric Plot: Archimedean Spiral')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axis('equal') # Equal scaling for x and y axes
plt.show()
```

### Even More Diverse Visualization Examples

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Scatter Plot with Regression Line Example

# Generate some correlated data
x_reg = np.random.rand(50) * 10
y_reg = 2 * x_reg + np.random.randn(50) * 5 + 5 # y = 2x + noise + intercept

plt.figure(figsize=(10, 6))
sns.regplot(x=x_reg, y=y_reg, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Stacked Bar Chart Example

# Sample data
categories = ['A', 'B', 'C', 'D']
subcategories = ['Type 1', 'Type 2', 'Type 3']
data_stacked = {
    'Type 1': np.random.randint(10, 50, 4),
    'Type 2': np.random.randint(5, 30, 4),
    'Type 3': np.random.randint(1, 20, 4)
}
df_stacked = pd.DataFrame(data_stacked, index=categories)

df_stacked.plot(kind='bar', stacked=True, figsize=(10, 7), colormap='viridis')
plt.title('Stacked Bar Chart of Categories by Type')
plt.xlabel('Category')
plt.ylabel('Value')
plt.xticks(rotation=0) # Keep category labels horizontal
plt.legend(title='Subcategory')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Polar Plot Example (Rose/Flower pattern)

# Generate data for a polar plot
theta = np.linspace(0, 2 * np.pi, 500) # Angle from 0 to 2*pi
r = np.sin(5 * theta) # A rose curve

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='polar')

ax.plot(theta, r, color='darkgreen', linewidth=2)
ax.set_title('Polar Plot: Rose Curve', va='bottom')
ax.grid(True)
plt.show()
```

### Creating a Legend

To add a legend to your plot, you typically do the following:

1.  **Label your plots**: When you call `plt.plot()` or similar plotting functions, use the `label` argument to give each element a descriptive name (e.g., `plt.plot(x, y, label='My Data')`).
2.  **Call `plt.legend()`**: After defining all your plot elements with labels, call `plt.legend()` to display the legend on the plot. Matplotlib automatically places it in a good location, or you can specify its position.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.plot(x, y2, label='Cosine Wave', color='red', linestyle='--')

plt.title('Plot with Legend Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add the legend using the labels provided in plt.plot()
plt.legend()

plt.grid(True)
plt.show()
```
